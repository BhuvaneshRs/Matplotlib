```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
```

Single plot

```python
plt.plot(x,np.sin(x))
```

Two Dimension plot

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

Change Line Style

```python
plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
```

```python
data=pd.read_excel("A:/Data Set/Sample - Superstore.xls")
```

```python
data
```

```python
plt.plot(data['Sales'],data['Profit'])
```

```python
plt.plot(data['Sales'],data['Profit'],"o")
```

```python
plt.subplot(3,1,1)
cat=data.groupby("Category")
fun=cat.get_group("Furniture")
off=cat.get_group("Office Supplies")
Tec=cat.get_group("Technology")
plt.plot(fun['Sales'],np.sin(fun['Sales']),"o")
plt.subplot(3,1,2)
plt.plot(off['Sales'],np.sin(off['Sales']),"o")
plt.subplot(3,1,3)
plt.plot(Tec['Sales'],np.sin(Tec['Sales']),"o")
```

```python
plt,a=plt.subplots(3)
a[0].plot(fun['Sales'],np.sin(fun['Sales']),"o")
a[1].plot(off['Sales'],np.sin(off['Sales']),"o")
a[2].plot(Tec['Sales'],np.sin(Tec['Sales']),"o")
```
